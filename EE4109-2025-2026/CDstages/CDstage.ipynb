## MOS CD stage

```python
CDstage = sl.makeCircuit("kicad/CDstages/CDstage.kicad_sch")
sl.img2html("CDstage.svg", 600)
```

### Gain

```python
CD_gain = sl.doLaplace(CDstage, pardefs="circuit").laplace
sl.eqn2html("Av_CDstage", CD_gain)
```

#### DC gain

```python
CD_DCgain = sp.limit(CD_gain, sl.ini.laplace, 0)
sl.eqn2html("AvDC_CDstage", CD_DCgain)
```

### Loop gain

```python
CD_loop_gain = sl.doLaplace(CDstage, pardefs="circuit", transfer="loopgain").laplace
sl.eqn2html("L_CDstage", sl.normalizeRational(CD_loop_gain))
```

#### DC loop gain, poles and zeros

```python
LPZ_CDstage = sl.doPZ(CDstage, pardefs="circuit", transfer="loopgain")
sl.pz2html(LPZ_CDstage)
```

1. The DC loop gain is low due to the bulk effect, with the bulk connected to the source, the DC loop 
2. equals the voltage gain factor of the CS stage.
3. The poles of the loop gain are real
4. The loop gain has one zero caused by c_dg

#### Output impedance

```python
CD_Zout = sl.doLaplace(CDstage, pardefs="circuit", source="I1").laplace
sl.eqn2html("Zo_CDstage", CD_Zout)
```

DC value

```python
CD_Rout = sp.limit(CD_Zout, sl.ini.laplace, 0)
sl.eqn2html("Ro_CDstage", CD_Rout)
```

If the source resistance is infinite and $c_{dg}$ equals zero, there's no longer feedback: $L=0$

```python
Zo = sp.limit(CD_Zout, sp.Symbol("R_s") , sp.oo)
Zo = sp.limit(Zo,      sp.Symbol("c_dg"), 0)
Ro = sp.limit(Zo,      sl.ini.laplace   , 0)
sl.eqn2html("R_o", sp.simplify(Ro))
```

## NMOS18 CD stage (bulk-source connected)

```python
N18CD_bulk = sl.makeCircuit("kicad/CDstages/N18CD_bulk.kicad_sch")
sl.img2html("N18CD_bulk.svg", 400)
```

### Feedback model

```python
par_defs = {"C_phz": 0, "W": "20u", "L": "2u", "ID": "100u", "R_s": "100k", "C_ell": "1p"}
N18CD_bulk.defPars(par_defs)

ag = sl.doLaplace(N18CD_bulk, pardefs="circuit", numeric=True, transfer="asymptotic")
lg = sl.doLaplace(N18CD_bulk, pardefs="circuit", numeric=True, transfer="loopgain")
sg = sl.doLaplace(N18CD_bulk, pardefs="circuit", numeric=True, transfer="servo")
dg = sl.doLaplace(N18CD_bulk, pardefs="circuit", numeric=True, transfer="direct")
gg = sl.doLaplace(N18CD_bulk, pardefs="circuit", numeric=True, transfer="gain")
sl.plotSweep("C18_fm_bulk", "C18 CD-stage bulk-source connected feedback model", [ag, lg, sg, dg, gg], 1, 1e3, 200, sweepScale="M")
sl.img2html("C18_fm_bulk.svg", 600)
```

```python
pzg = sl.doPZ(N18CD_bulk, pardefs="circuit", numeric=True)
sl.pz2html(pzg)
```

### Frequency compensation

```python
pzl   = sl.doPZ(N18CD_bulk, pardefs="circuit", transfer="loopgain", numeric=True)
L_DC  = pzl.DCvalue                            # Value of DC loop gain
poles = pzl.poles                              # List with poles of the loop gain in rad/s
p_1   = poles[0]                               # Pole in rad/s
p_2   = poles[1]                               # Pole in rad/s
B_w   = sp.sqrt((1 - L_DC) * p_1 * p_2)        # Second-order bandwidth
z     = -B_w**2/(sp.sqrt(2) * B_w + p_1 + p_2) # Phantom-zero frequency for MFM
Cphz  = -1/(z * N18CD_bulk.getParValue("R_s")) # Compensation capacitance
sl.eqn2html("C_phz", Cphz)
```

```python
N18CD_bulk.defPar("C_phz", Cphz)
pzg_comp = sl.doPZ(N18CD_bulk, pardefs="circuit", numeric=True)
sl.pz2html(pzg_comp)
```

```python
gm = N18CD_bulk.getParValue("g_m_X1", numeric=True)
step_dict = {"params": "g_m_X1",
             "method": "lin",
             "start":  0,
             "stop":   gm,
             "num": 100}
RL = sl.doPoles(N18CD_bulk, pardefs="circuit", transfer="servo", stepdict=step_dict, numeric=True)
PA = sl.doPoles(N18CD_bulk, pardefs="circuit", transfer="asymptotic", numeric=True)
PL = sl.doPoles(N18CD_bulk, pardefs="circuit", transfer="loopgain", numeric=True)
ZL = sl.doZeros(N18CD_bulk, pardefs="circuit", transfer="loopgain", numeric=True)
sl.ini.axis_height = 6
sl.plotPZ("RL_comp", "Root locus", [PA, PL, ZL, RL], xmin=-50, xmax=0, ymin=-25, ymax=25, xscale="M", yscale="M")
sl.img2html("RL_comp.svg", 600)
```

```python
step_dict = {"params": "C_phz",
             "method": "lin",
             "start":  0,
             "stop":   sp.N(2*Cphz),
             "num": 10}
RL_phz = sl.doPoles(N18CD_bulk, pardefs="circuit", transfer="servo", stepdict=step_dict, numeric=True)
sl.plotPZ("RL_comp_phz", "Root locus", RL_phz, xmin=-50, xmax=0, ymin=-25, ymax=25, xscale="M", yscale="M")
sl.img2html("RL_comp_phz.svg", 600)
```

### Automatic compensation (order = -2)

```python
N18CD_bulk.defPar("C_phz", 0)
lg_uncomp   = sl.doLaplace(N18CD_bulk, pardefs="circuit", transfer="loopgain", numeric=True).laplace
# Find the number of dominant poles and the bandwidth in rad/s
sl.ini.hz   = False
servo_data  = sl.findServoBandwidth(lg_uncomp)
order       = servo_data['lpo']
Bw          = servo_data['lpf']
if order == -2:
    # Obtain the poles and calculate the sum of the poles
    poles       = sl.doPoles(N18CD_bulk, pardefs="circuit", transfer="loopgain", numeric=True).poles
    n_dom_poles = int(sp.Abs(order))
    sum_n_poles = 0
    for i in range(n_dom_poles):
        sum_n_poles += poles[i]
    # Calculate the zero and the value of C_phz
    z = Bw**2/(sp.sqrt(2) * Bw + sum_n_poles)
    Cphz  = 1/(z * N18CD_bulk.getParValue("R_s")) # Compensation capacitance
    sl.eqn2html("C_phz", Cphz)
else:
    raise NotImplementedError("Oops; expected order -2, got %s!"%(str(order)))
```
