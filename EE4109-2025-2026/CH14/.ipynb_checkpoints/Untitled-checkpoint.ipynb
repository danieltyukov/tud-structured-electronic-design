# Dual CS-stage controller amplifiers

```python
import SLiCAP as sl
import sympy as sp
sl.initProject("Dual CS-stage controller amplifiers", notebook=True)
```

## Voltage amplifier
We will study the asymptotic-gain of negative feedback voltage amplifier using two unbalanced CS stages as controller. We will do this for four cases:
1. The signal source is connected to the signal reference (ground)
   - The transconductance factor of the first stage is selected as loop gain reference
   - The transconductance factor of the second stage is selected as loop gain reference
2. The signal source is floating; the positions of the controller input and the signal source are swapped w.r.t. the previous case.
   - The transconductance factor of the first stage is selected as loop gain reference
   - The transconductance factor of the second stage is selected as loop gain reference

## Configuration with grounded signal source

```python
Vamp1 = sl.makeCircuit("kicad/dualStageCScontroller/VampDualStageCScontroller_1.kicad_sch")
sl.img2html("VampDualStageCScontroller_1.svg", 400)
```

### Loop gain reference in the first stage

```python
AG11 = sl.doLaplace(Vamp1, lgref="Gm_M1", transfer="asymptotic").laplace
A_foo11 = sp.Symbol("A_foo11")
sl.eqn2html(A_foo11, AG11)
```

```python
expr = sp.Limit(A_foo11, sp.Symbol("c_gs2"), 0)
sl.eqn2html(expr, expr.subs(A_foo11, AG11).doit())
```

### Loop gain reference in the second stage

```python
AG12 = sl.doLaplace(Vamp1, lgref="Gm_M2", transfer="asymptotic").laplace
A_foo12 = sp.Symbol("A_foo12")
sl.eqn2html(A_foo12, AG12)
```

```python
expr = sp.Limit(sp.Limit(A_foo12, sp.Symbol("g_b1"), 0), sp.Symbol("g_o1"), 0)
sl.eqn2html(expr, expr.subs(A_foo12, AG12).doit())
```

### Conclusions
- With the loop gain reference selected in the first stage, the servo function accurately describes the discrepancy between the ideal gain and the actual gain at frequencies at which the influence of the input capacitance of the second stage on the transfer can be ignored.
- With the loop gain reference selected in the second stage, the servo function accurately describes the discrepancy between the ideal gain and the actual gain, if the influence of the bulk transconductance and the output conductance of the first stage can be ignored. The effect of the bulk tranconductance on the transfer can be eliminated by connecting the bulk to the source. The effect of the output conductance on the transfer can be reduced by increasing the channel length or using a cascoded input stage.

## Explanation

```python
cir = sl.makeCircuit("kicad/dualStageCScontroller/explanation.kicad_sch")
sl.img2html("explanation.svg", 400)
```

We will evaluate the controller input voltage $-V_{out}$ with the second stage loop gain reference
replaced with a nullor. 

```python
V_in = sl.doLaplace(cir, transfer=None).laplace
sl.eqn2html("V_in", V_in)
```

This expresion shows that in case of a grounded signal source the controller input voltage does not become zero with the second stage loop gain reference approaching infinity. In this case the first stage needs to produce a current that flows through the first stage output resistance to create approximately $V_s$. This requires a non-zero drive voltage for the input stage.

With a grounded controller input it follows that $V_{in}=0$.
