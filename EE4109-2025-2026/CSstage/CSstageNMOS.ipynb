# CS stage

```python
import SLiCAP as sl
sl.initProject("CSstage", notebook=True)
%store -r model
```

## Voltage gain and transadmittance

```python
netlist = sl.makeCircuit("kicad/CSstage/CSstageNMOS_SPICE_Av-Yt.kicad_sch", language="SPICE")
```

### Biasing and checking operating point

```python
sim_cmd = "OP"
names   = {"I_D"   : "@M3[id]",
           "V_out" : "V(out)",
           "V_load": "V(load)"}
OPinfo = sl.ngspice2traces("cir/CSstageNMOS_SPICE_Av-Yt", sim_cmd, names)
sl.backAnnotateSchematic("kicad/CSstage/CSstageNMOS_SPICE_Zt-Ai.kicad_sch", OPinfo)
sl.img2html("CSstageNMOS_SPICE_Av-Yt.svg", 600)
```

## AC analysis: small-signal dynamic behavior; frequency-domain analysis

```python
sim_cmd = "AC DEC 20 1MEG 100G"
names   = {"V_out": "V(out)", "ID": "I(V3)", "V_load": "V(load)", "I_load": "I(V4)"}
mag, phase, x_name, x_units  = sl.ngspice2traces("cir/CSstageNMOS_SPICE_Av-Yt", sim_cmd, names, traceType='magPhase')
Av = sl.selectTraces(mag, ["V_out", "V_load"])
sl.plot("CSstageNMOS_AC_Av", "Magnitude voltage gain", "log", Av, xName=x_name, xUnits=x_units, xScale="M", 
        yName="$A_v$")
sl.img2html("CSstageNMOS_AC_Av.svg", 400)
```

```python
AvPhase = sl.selectTraces(phase, ["V_out", "V_load"])
sl.plot("CSstageNMOS_AC_Av_phase", "Phase voltage gain", "semilogx", AvPhase, xName=x_name, xUnits=x_units, xScale="M", 
        yName="$arg(A_v)$", yUnits="deg")
sl.img2html("CSstageNMOS_AC_Av_phase.svg", 400)
```

```python
Y_t = sl.selectTraces(mag, ["ID", "I_load"])
sl.plot("CSstageNMOS_AC_Yt", "Magnitude transadmittance", "semilogx", Y_t, xName=x_name, xUnits=x_units, xScale="M", 
        yScale="u", yName="$Y_t$", yUnits="S", yLim=[80, 100])
sl.img2html("CSstageNMOS_AC_Yt.svg", 400)
```

```python
Y_tPhase = sl.selectTraces(phase, ["ID", "I_load"])
sl.plot("CSstageNMOS_AC_Yt_phase", "Magnitude transadmittance", "semilogx", Y_tPhase, xName=x_name, xUnits=x_units, xScale="M", 
        yName="$arg(Y_t)$", yUnits="deg")
sl.img2html("CSstageNMOS_AC_Yt_phase.svg", 400)
```

## Transient analysis: nonlinear time-variant dynamic behavior; time-domain analysis

```python
sim_cmd = "TRAN 10n 5u"
Out_Av, x_name, x_units  = sl.ngspice2traces("cir/CSstageNMOS_SPICE_Av-Yt", sim_cmd, names)
Vload_Av = sl.selectTraces(Out_Av, ["V_load"])
sl.plot("CSstageNMOS_TRAN_Vload_Av", "$V_{load}$", "lin", Vload_Av, xName=x_name, xUnits=x_units, xScale="u", 
        yName="$V_{load}$", yUnits="V")
sl.img2html("CSstageNMOS_TRAN_Vload_Av.svg", 400)
```

```python
Iload_Av = sl.selectTraces(Out_Av, ["I_load"])
sl.plot("CSstageNMOS_TRAN_ID_Av", "$I_{V3}$", "lin", Iload_Av, xName=x_name, xUnits=x_units, xScale="u", 
        yName="$I_{V3}$", yScale="u", yUnits="A")
sl.img2html("CSstageNMOS_TRAN_ID_Av.svg", 400)
```

## Transimpedance and current gain

```python
netlist = sl.makeCircuit("kicad/CSstage/CSstageNMOS_SPICE_Zt-Ai.kicad_sch", language="SPICE")
```

### Biasing and checking operating point

```python
sim_cmd = "OP"
names   = {"I_D"   : "@M3[id]",
           "V_out" : "V(out)",
           "V_load": "V(load)"}
OPinfo = sl.ngspice2traces("cir/CSstageNMOS_SPICE_Zt-Ai", sim_cmd, names)
sl.backAnnotateSchematic("kicad/CSstage/CSstageNMOS_SPICE_Zt-Ai.kicad_sch", OPinfo)
sl.img2html("CSstageNMOS_SPICE_Zt-Ai.svg", 600)
```

## AC analysis: small-signal dynamic behavior; frequency-domain analysis

```python
sim_cmd = "AC DEC 20 1MEG 100G"
names   = {"V_out": "V(out)", "ID": "I(V3)", "V_load": "V(load)", "I_load": "I(V4)"}
mag, phase, x_name, x_units  = sl.ngspice2traces("cir/CSstageNMOS_SPICE_Zt-Ai", sim_cmd, names, traceType='magPhase')
Z_t = sl.selectTraces(mag, ["V_out", "V_load"])
Zt_spice = sl.plot("CSstageNMOS_AC_Zt", "magnitude transimpedance", "log", Z_t, xName=x_name, xUnits=x_units, xScale="M", 
                    yName="$Z_t$", yUnits="$\\Omega$")
sl.img2html("CSstageNMOS_AC_Zt.svg", 400)
```

```python
Z_t_phase = sl.selectTraces(phase, ["V_out", "V_load"])
sl.plot("CSstageNMOS_AC_Zt_phase", "phase transimpedance", "semilogx", Z_t_phase, xName=x_name, xUnits=x_units, xScale="M", 
        yName="$arg(Z_t)$", yUnits="$deg$")
sl.img2html("CSstageNMOS_AC_Zt_phase.svg", 400)
```

```python
A_i = sl.selectTraces(mag, ["ID", "I_load"])
sl.plot("CSstageNMOS_AC_Ai", "magnitude current gain", "log", A_i, xName=x_name, xUnits=x_units, xScale="M", 
        yName="$A_i$")
sl.img2html("CSstageNMOS_AC_Ai.svg", 400)
```

```python
A_i_phase = sl.selectTraces(phase, ["ID", "I_load"])
sl.plot("CSstageNMOS_AC_Ai_phase", "phase current gain", "semilogx", A_i_phase, xName=x_name, xUnits=x_units, xScale="M", 
        yName="$arg(A_i)$", yUnits="deg")
sl.img2html("CSstageNMOS_AC_Ai_phase.svg", 400)
```

## Transient analysis: nonlinear time-variant dynamic behavior; time-domain analysis

```python
sim_cmd = "TRAN 10n 5u"
Out, x_name, x_units  = sl.ngspice2traces("cir/CSstageNMOS_SPICE_Zt-Ai", sim_cmd, names)
Vload = sl.selectTraces(Out, ["V_load"])
sl.plot("CSstageNMOS_TRAN_Zt", "$V_{load}$", "lin", Vload, xName=x_name, xUnits=x_units, xScale="u", 
        yName="$V$", yUnits="V")
sl.img2html("CSstageNMOS_TRAN_Zt.svg", 400)
```

```python
Iload = sl.selectTraces(Out, ["I_load"])
sl.plot("CSstageNMOS_TRAN_Ai", "$I_{V3}$", "lin", Iload, xName=x_name, xUnits=x_units, xScale="u", 
        yName="$I_{V3}$", yScale="u", yUnits="A")
sl.img2html("CSstageNMOS_TRAN_Ai.svg", 400)
```
