```python
import SLiCAP as sl
```

```python
# Define a project
sl.initProject('NMOS EKV plots', notebook=True)
```

```python
fileName = 'mosEKVplots.cir'
cir = sl.makeCircuit(fileName)     # Checks and defines the local circuit object
```

```python
sl.netlist2html(fileName)
```

```python
# Print data of elements of expanded network
sl.elementData2html(cir)
```

```python
# Print the circuit parameters
sl.params2html(cir)
```

```python
# Define step parameters
step_dict = {"params": "V_G",
             "method": "lin",
             "start" :  0.001,
             "stop"  :   1.8,
             "num"   : 10}
result = sl.doParams(cir, stepdict=step_dict)
```

```python
fig_Ids_Vds = sl.plotSweep('IdsVds', '$I_{ds}(V_{ds})$', result, 0, 1.8, 50, sweepVar= 'V_D', xUnits = 'V', yVar = 'I_DS_X1', yScale = 'u', yUnits = 'A', funcType = 'param')
sl.img2html('IdsVds.svg', 600)
```

```python
# All other plots will be made with V_GS as sweep parameter and V_DS as step parameter.
step_dict["params"] = "V_D"
result = sl.doParams(cir, stepdict=step_dict)
```

```python
fig_Ids_Vgs = sl.plotSweep('IdsVgs', '$I_{ds}(V_{gs})$', result, 0, 1.8, 50, sweepVar= 'V_G', xUnits = 'V', yVar = 'I_DS_X1', yScale = 'u', yUnits = 'A', funcType = 'param')
sl.img2html('IdsVgs.svg', 600)
```

```python
fig_gm_Ids = sl.plotSweep('gmIds', '$g_m(I_{ds})$', result, 0, 1.8, 50, sweepVar= 'V_G', xVar = 'I_DS_X1', xScale = 'u', xUnits = 'A', yVar = 'g_m_X1', yScale = 'u', yUnits = 'S', funcType = 'param')
sl.img2html('gmIds.svg', 600)
```

```python
fig_fT_Ids = sl.plotSweep('fTIds', '$f_{T}(I_{ds})$', result, 0, 1.8, 50, sweepVar= 'V_G', xVar = 'I_DS_X1', xScale = 'u', xUnits = 'A', yVar = 'f_T_X1', yScale = 'G', yUnits = 'Hz', funcType = 'param')
sl.img2html('fTIds.svg', 600)
```

```python
fig_CissVg = sl.plotSweep('CissVg', '$c_{iss}(V_{gs})$', result, 0, 1.8, 50, sweepVar= 'V_G', xScale = '', xUnits = 'V', yVar = 'c_iss_X1', yScale = 'f', yUnits = 'F', funcType = 'param')
sl.img2html('CissVg.svg', 600)
```
