# Differential-pair CS (anti-seris CS)

```python
import SLiCAP as sl
sl.initProject("Balanced stages", notebook=True)
```

## Test setup transadmittance

```python
diffPair = sl.makeCircuit("kicad/balancing/diffPair.kicad_sch", language="SPICE")
sl.img2html("diffPair.svg", 450)
```

### Odd-order DM transfer

```python
sim_cmd = "DC V3 -500m 500m 1m"
step_cmd = "Iss lin 0.5u 50u 3"
names    = {"I_diff": "(I(V4)-I(V5))/(I(V4)+I(V5))"}
DC, x_name, x_units = sl.ngspice2traces("cir/diffPair", sim_cmd, names, stepCmd=step_cmd)
sl.plot("diffPar_dm", "Relative differential-mode output current", "lin", 
        DC, xName=x_name, xUnits="V", yName="$\\frac{I_{diff}}{I_{ss}}$")
sl.img2html("diffPar_dm.svg", 600)
```

### Even-order CM qantities

```python
names    = {"V_ss": "V(ss)"}
DC, x_name, x_units = sl.ngspice2traces("cir/diffPair", sim_cmd, names, stepCmd=step_cmd)
sl.plot("diffPar_cm", "Common-mode source voltage", "lin", 
        DC, xName=x_name, xUnits="V", yName="$V_{ss}$", yUnits="V")
sl.img2html("diffPar_cm.svg", 600)
```

## Test circuit voltge output

### Common-mode biasing (local feedback concept)

```python
diffPairVout1 = sl.makeCircuit("kicad/balancing/diffPairVout1.kicad_sch", language="SPICE")
sim_cmd = "OP"
names   = {"V_outP": "V(outP)", 
           "V_outN": "V(outN)", 
           "I_D1": "@M1[id]", 
           "I_D2": "@M2[id]"}
opinfo  = sl.ngspice2traces("cir/diffPairVout1", sim_cmd, names)
sl.backAnnotateSchematic("kicad/balancing/diffPairVout1.kicad_sch", opinfo)
sl.img2html("diffPairVout1.svg", 900)
```

```python
sim_cmd = "DC V3 -100m 100m 1m"
step_cmd = "Iss lin 0.5u 50u 3"
names    = {"V_diff": "V(outP)-V(outN)"}
DC, x_name, x_units = sl.ngspice2traces("cir/diffPairVout1", sim_cmd, names, stepCmd=step_cmd)
sl.plot("diffPairVout1_dm", "Differential-mode output voltage", "lin", 
        DC, xName=x_name, xUnits="V", yName="$V_{diff}$", yUnits="V")
sl.img2html("diffPairVout1_dm.svg", 600)
```
