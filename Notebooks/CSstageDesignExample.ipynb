# CS stage noise design automation

```python
import SLiCAP as sl
from SLiCAPmosNoise import process, specObject, doMOSnoiseDesign
sl.initProject("MOSdesign", notebook=True)
```

## Define the circuit

```python
# Enter the name of the kicad schematic of the feedback amplifier with nullor
fileName  = "Trimp"
kicad_sch = "kicad/{}/{}.kicad_sch".format(fileName, fileName)
```

## Define the nosie performance specifications

```python
specs = specObject()
specs.noiseRMS = 100e-6  # Total RMS output noise budget over frequency range
specs.f_min    = 1       # Lower limit of frequency range
specs.f_max    = 1e6     # Upper limit of frequency range
specs.ID_spec  = 1e-3    # Current budget; always positive, also for P-channel
specs.IC       = 5       # IC for noise design according to specs
specs.IC_max   = 16      # IC for minimum noise (calculated from c_issOpt @ fT @ IC_max)
specs.gmID     = 10      # gmID for noise design according to specs
                         # will be overruled by specs.IC
specs.gmID_min = None   # gmID for minimum noise (maximum fT)
                         # will be overruled by specs.IC_max
```

## Define the technology specifications
Can be a list with multiple technology models

```python
models = []

model         = process()
model.name    = "MN18_noisyNullor"
model.channel = "N"
model.IG      = 0
model.L_min   = 0.18e-6
model.L_max   = 50e-6
model.W_min   = 0.18e-6
model.W_max   = 10e-3

models.append(model)

model         = process()
model.name    = "MP18_noisyNullor" 
model.channel = "P"
model.IG      = 0
model.L_min   = 0.18e-6
model.L_max   = 50e-6
model.W_min   = 0.18e-6
model.W_max   = 10e-3

models.append(model)
```

## Obtain the results

```python
all_results = doMOSnoiseDesign(kicad_sch, specs, models, printAll=True)
sl.img2html("{}.svg".format(fileName), 400)
```
